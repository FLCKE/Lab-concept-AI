```python
import pandas as pd

pd.set_option('display.max_columns', None)
df = pd.read_csv("application_train.csv")
df.head()
```

Verifier s'il ya des donnée manquante dans la data et les supprimer

```python

df.isnull().any()
df.isnull().sum()
df[df.isnull().any(axis=1)]
df = df.dropna() 
df.isnull().sum()
print(df.dtypes)
```

verifier si les colonne ne contient pas des types de donnés diférents 

```python
df.describe()
```

```python
df['TARGET'].value_counts(normalize=True)
```

verifier si le nombre d'enfant varie selon le rembourser et le non rembourser 

```python
import matplotlib.pyplot as plt


prop = df.groupby('CNT_CHILDREN')['TARGET'].value_counts(normalize=True).unstack(fill_value=0).sort_index()

for c in [0, 1]:
    if c not in prop.columns:
        prop[c] = 0
prop = prop[[0, 1]]

ax = prop.plot(kind='bar', stacked=True, color=['#4CAF50', '#F44336'], figsize=(9,5))
ax.set_xlabel("Nombre d'enfants (CNT_CHILDREN)")
ax.set_ylabel("Proportion")
ax.set_title("Répartition du TARGET (0 = remboursé, 1 = non remboursé) par nombre d'enfants")
ax.legend(title="TARGET", labels=["Remboursé (0)", "Non remboursé (1)"])

for p in ax.patches:
    h = p.get_height()
    if h > 0:
        x = p.get_x() + p.get_width() / 2
        y = p.get_y() + h / 2
        ax.annotate(f"{h:.0%}", (x, y), ha="center", va="center", color="white", fontsize=9)

plt.tight_layout()
plt.show()
```

De ce graqhe on peut comprendre que le nombre d'enfants d'influence pas vraiment sur le retard ou le non remboursement 

Je vais verifier si le genre influence sur le remboursement ou non 

```python
import matplotlib.pyplot as plt


prop = df.groupby('CODE_GENDER')['TARGET'].value_counts(normalize=True).unstack(fill_value=0).sort_index()

for c in [0, 1]:
    if c not in prop.columns:
        prop[c] = 0
prop = prop[[0, 1]]

ax = prop.plot(kind='bar', stacked=True, color=['#4CAF50', '#F44336'], figsize=(9,5))
ax.set_xlabel("Genre (CODE_GENDER)")
ax.set_ylabel("Proportion")
ax.set_title("Répartition du TARGET (0 = remboursé, 1 = non remboursé) par genre")
ax.legend(title="TARGET", labels=["Remboursé (0)", "Non remboursé (1)"])

for p in ax.patches:
    h = p.get_height()
    if h > 0:
        x = p.get_x() + p.get_width() / 2
        y = p.get_y() + h / 2
        ax.annotate(f"{h:.0%}", (x, y), ha="center", va="center", color="white", fontsize=9)

plt.tight_layout()
plt.show()
```

on remarque que le nombre de personne de sexe masculin n'ayant pas remboursé est légerement supérieur a celui de sexe féminin mais je ne peux pas tirer de conclusion sur cela car ce n'est qu'une legere différence de 2%
